```python
yearsales=df.groupby("year")["rating"].count().sort_values(ascending=False)[:10]
```

```python
yearsales
```

```python
yearsales.index
```

```python
yearsales.values
```

```python
plt.figure(figsize=(14,6))
plt.bar(yearsales.index,yearsales.values,width=0.4)
plt.xticks(yearsales.index)
plt.yticks(np.arange(0,330000,20000))
plt.xlabel("year")
plt.ylabel("unit sold")
plt.title("year wise unit sold")
plt.show()
```

```python
categorysales=df.groupby("category")["rating"].count().sort_values(ascending=False)
categorysales
```

```python
color=["r","g","y","pink","m","b","k","yellowgreen","purple","orange"]
plt.barh(categorysales.index,categorysales.values,color=color)

plt.show()
```

```python
model_attr_sales=df.groupby("model_attr")["rating"].count().sort_values(ascending=False)
model_attr_sales
```
